## `SBI stock Analysis`

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('SBIN_Data.csv')
data.head(20)
```

```python
data['DateTime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d-%m-%Y %H:%M:%S')
data.head()
```

```python
# Group the data by 'Time'
grouped = data.groupby(data['DateTime'].dt.time)
# data.head()
grouped.head()
```

```python
def volume_rank(group):
    last_day = group['DateTime'].dt.date.max()
    last_5_day = group[group['DateTime'].dt.date.between(last_day - pd.Timedelta(days=5), last_day)]
    volume = last_5_day.groupby(last_5_day['DateTime'].dt.time)['Volume']
    rank = volume.rank(method='first' , ascending=False)
    return rank
```

```python

data['Rank'] = grouped.apply(volume_rank).reset_index(level=0, drop=True)
```
